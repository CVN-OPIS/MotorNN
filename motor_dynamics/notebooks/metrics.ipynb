```python
import pickle as pkl 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
fin = open('../../../results/Data8Feb2019.pkl','rb')
data = pkl.load(fin)
fin.close()
```

```python
plt.figure(figsize=(30,10))
plt.plot(data['time'],data['speed_true'])
plt.plot(data['time'],data['speed_pred'])
```

```python
plt.figure(figsize=(30,10))
plt.plot(data['time'],data['torque_true'])
plt.plot(data['time'],data['torque_pred'])
```
