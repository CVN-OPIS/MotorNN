```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import cv2
import csv
```

```python
a30_c1 = csv.reader(open('../datasets/ann30_curr1.log','rb'))
a30_c2 = csv.reader(open('../datasets/ann30_curr2.log','rb'))
a30_t = csv.reader(open('../datasets/ann30_torque.log','rb'))
```

```python
ted = 7.5
trd = 40
```

```python
c1d_tr = []
c1d_te = []
for row in a30_c1:
    c1d_tr.append(float(row[1])/trd)
    c1d_te.append(float(row[2])/ted)
```

```python
plt.plot(c1d_tr, label='train loss')
plt.plot(c1d_te, label='test loss')
plt.legend()
plt.title('ANN Current1, w=30')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.savefig('ann30_curr1.pdf', bbox_inches='tight')
```

```python
c2d_tr = []
c2d_te = []
for row in a30_c2:
    c2d_tr.append(float(row[1])/trd)
    c2d_te.append(float(row[2])/ted)
```

```python
plt.plot(c2d_tr, label='train loss')
plt.plot(c2d_te, label='test loss')
plt.legend()
plt.title('ANN Curren2, w=30')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.savefig('ann30_curr2.pdf', bbox_inches='tight')
```

```python
td_tr = []
td_te = []
for row in a30_t:
    td_tr.append(float(row[1])/trd)
    td_te.append(float(row[2])/ted)
```

```python
plt.plot(td_tr, label='train loss')
plt.plot(td_te, label='test loss')
plt.legend()
plt.title('ANN Torque, W=30')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.savefig('ann30_torque.pdf', bbox_inches='tight')
```
