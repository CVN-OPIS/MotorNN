```python
from math import *
%matplotlib inline
import matplotlib.pyplot as plt

from scipy.signal import sawtooth, square, triang, bessel
import numpy as np
```

```python
t_triag = [x * 0.005 for x in range(200)]
triag = triang(len(t_triag))
t_saw = [x * 0.005 for x in range(200,400)]
saw =  sawtooth(5 * np.asarray(t_saw))
```

```python
plt.plot(t_triag, triag)
plt.plot(t_saw, saw)
```
