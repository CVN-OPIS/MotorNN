```python
import numpy as np
from math import *

%matplotlib inline
import matplotlib.pyplot as plt

import scipy.io as sio
from scipy.fftpack import dct, idct

from numpy.linalg import norm
```

```python
def hanning(L):
    return np.asarray([0.5* (1 - cos(x)) for x in (2 * pi * np.asarray(list(range(1,L+1)))) / (L+1)])
```

```python
L = 16
dcttype = 4
torque = sio.loadmat('../../datasets/CS2018_12_14/Torque.mat')
time = sio.loadmat('../../datasets/CS2018_12_14/t.mat')['t']
y = torque['Torque']
N = y.shape[0]
```

```python
w = hanning(L)
c = np.zeros((N-L+1,L))
```

```python
for n in range(0,N-L+1):
    Y = y[n:n+L]
    c[n, :] = w * Y.flatten()
```

```python
c = dct(c, type=dcttype, n=L, norm='ortho')
```

```python
st = 12
en = 24
start = int(st / 0.005)
end = int(en / 0.005)

f = plt.figure(figsize=(25,15))

f.add_subplot(4,4,1)
plt.plot(time[start:end], c[start:end,0])

f.add_subplot(4,4,2)
plt.plot(time[start:end], c[start:end,1])

f.add_subplot(4,4,3)
plt.plot(time[start:end], c[start:end,2])

f.add_subplot(4,4,4)
plt.plot(time[start:end], c[start:end,3])

f.add_subplot(4,4,5)
plt.plot(time[start:end], c[start:end,4])

f.add_subplot(4,4,6)
plt.plot(time[start:end], c[start:end,5])

f.add_subplot(4,4,7)
plt.plot(time[start:end], c[start:end,6])

f.add_subplot(4,4,8)
plt.plot(time[start:end], c[start:end,7])

f.add_subplot(4,4,9)
plt.plot(time[start:end], c[start:end,8])

f.add_subplot(4,4,10)
plt.plot(time[start:end], c[start:end,9])

f.add_subplot(4,4,11)
plt.plot(time[start:end], c[start:end,10])

f.add_subplot(4,4,12)
plt.plot(time[start:end], c[start:end,11])

f.add_subplot(4,4,13)
plt.plot(time[start:end], c[start:end,12])

f.add_subplot(4,4,14)
plt.plot(time[start:end], c[start:end,13])

f.add_subplot(4,4,15)
plt.plot(time[start:end], c[start:end,14])

f.add_subplot(4,4,16)
plt.plot(time[start:end], c[start:end,15])
```

```python
N = c.shape[0] + L - 1
invc = idct(c, type=dcttype, n=L, norm='ortho').T
cf = np.fliplr(invc)
yr = np.zeros((N))
```

```python
for nn in range(N-L,-L,-1):
    d = np.diag(cf, nn)
    if nn >= 0:
        ww = w[:d.shape[0]]
    else:
        ww = w[-nn:L]
      
    yr[N-L-nn] = np.dot(ww, d / norm(ww)**2)
```

```python
norm(y.flatten()-yr)
```

```python
plt.plot(time[start:end], yr[start:end])
```
